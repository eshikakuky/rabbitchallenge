```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegressionCV
from sklearn.metrics import confusion_matrix
from sklearn.metrics import recall_score, accuracy_score
#from sklearn.decomposition import PCA

from sklearn.cluster import KMeans
from sklearn.neighbors import KNeighborsClassifier


import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# データの読み込み
cancer_df = pd.read_csv('/content/drive/My Drive/study_ai_ml_google/data/cancer.csv')
```

```python
# ‘Unnamed’列は欠損データのみのため削除　
cancer_df.drop('Unnamed: 32', axis=1, inplace=True)
```

・diagnosis: 診断結果 (良性がB / 悪性がM)
・説明変数は3列以降、目的変数を2列目としロジスティック回帰で分類

```python
# 目的変数の抽出
cancer_df['diag'] = cancer_df.diagnosis.apply(lambda d: 1 if d == 'M' else 0)
y = cancer_df['diag'] 
```

```python
# 説明変数の抽出    'radius_mean'から‘fractal_dimension_worst’までの３０列
X = cancer_df.loc[:, 'radius_mean':'fractal_dimension_worst']
```

```python
y
```

```python
X
```

```python
X.info()
```

```python
# 学習用とテスト用でデータを分離
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# 標準化
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### ロジスティック回帰で学習

```python
# ロジスティック回帰で学習
#logistic = LogisticRegressionCV(cv=10, random_state=0)
logistic = LogisticRegressionCV(cv=10, random_state=0, max_iter=200)
logistic.fit(X_train_scaled, y_train)

# 検証
#print('Train score: {:.3f}'.format(logistic.score(X_train_scaled, y_train)))
#print('Test score: {:.3f}'.format(logistic.score(X_test_scaled, y_test)))
print('Confustion matrix:\n{}'.format(confusion_matrix(y_true=y_test, y_pred=logistic.predict(X_test_scaled))))

#                 Predicted
#                 Negative  Positive
#Actual Negative     TN        FP
#       Positive     FN        TP


# accuracy  正解率の算出
print('accuracy score:{:.4f}'.format( accuracy_score(y_true=y_test, y_pred=logistic.predict(X_test_scaled))))
# recall 再現率の算出
print('reacll score:{:.4f}'.format( recall_score(y_true=y_test, y_pred=logistic.predict(X_test_scaled))))
```

   　※ロジスティクス回帰による分類予測　　　　正解率：　0.972　　再現率：　0.943


# k-近傍法による分類

```python
# k-NN モデルの生成
knc = KNeighborsClassifier(n_neighbors=5).fit(X_train, y_train)
```

```python
pred_knc_y = knc.predict(X_test)
```

```python
print('Confustion matrix:\n{}'.format(confusion_matrix(y_true=y_test, y_pred=pred_knc_y)))

#                 Predicted
#                 Negative  Positive
#Actual Negative     TN        FP
#       Positive     FN        TP

# accuracy  正解率の算出
print('accuracy score:{:.4f}'.format( accuracy_score(y_true=y_test, y_pred=pred_knc_y)))
# recall 再現率の算出
print('reacll score:{:.4f}'.format( recall_score(y_true=y_test, y_pred=pred_knc_y)))
```

   　※k-近傍法による分類予測　　　　正解率：　0.937　　再現率：　0.925

  

# k-平均法による分類

```python
# k-means モデルの生成
model_KM = KMeans(n_clusters=2, random_state=0)
```

```python
model_KM.fit(X_train)
pred_kms_y = model_KM.predict(X_test)
```

```python
print('Confustion matrix:\n{}'.format(confusion_matrix(y_true=y_test, y_pred=pred_kms_y)))

#                 Predicted
#                 Negative  Positive
#Actual Negative     TN        FP
#       Positive     FN        TP


# accuracy  正解率の算出
print('accuracy score:{:.4f}'.format( accuracy_score(y_true=y_test, y_pred=pred_kms_y)))
# recall 再現率の算出
print('reacll score:{:.4f}'.format( recall_score(y_true=y_test, y_pred=pred_kms_y)))
```

   　※k-平均法による分類予測　　　　正解率：　0.846　　再現率：　0.585
